```python
import pandas as pd
import pandas.io.data as web # Package and modules for importing data   
import datetime
# from pandas_datareader import data, wb
 
# Stock prices over the past year, starting at January 1, 2016
start = datetime.datetime(2016,1,1)
end = datetime.date.today()
 
# Apple stock data
appl = web.DataReader("AAPL",  # stock ticker
                       "yahoo", # data source - Yahoo Finance
                       start,   # start date
                       end      # end date
                      )
 
type(appl)
```

Writing and reading dataframe to JSON

http://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_json.html
    
http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_json.html

```python
df = appl

print( df.reset_index().to_json() )
```

```python
print( df.reset_index().to_json(orient='split') )
```

```python
df = appl

json_str = df.reset_index().to_json(orient='records')

print( type(json_str) )

print( df.reset_index().to_json(orient='records') )
```
